```python
"""
NetworkX is a Python package for the creation, manipulation, 
and study of the structure, dynamics
and functions of complex networks.
"""
import matplotlib.pyplot as plt
import networkx as nx

edges = [['A', 'B'], ['B', 'C'], ['B', 'D']] #Setting the edges
G = nx.Graph() #Displaying the Graphs using networkX
G.add_edges_from(edges)
pos = nx.spring_layout(G)
plt.figure()
nx.draw(
    G, pos, edge_color='black', width=1, linewidths=1,
    node_size=500, node_color='pink', alpha=0.9,
    labels={node: node for node in G.nodes()}
) #We Draw the figure using nx properties
nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels={('A', 'B'): 'AB', 
                 ('B', 'C'): 'BC', 
                 ('B', 'D'): 'BD'},
    font_color='red'
)
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.DiGraph() #We display the graph
G.add_edges_from([(1, 2), (1, 3), (2, 3)]) #Setting the edges.

pos = nx.spring_layout(G)
nx.draw_networkx(G, pos)
edge_labels = dict([((n1, n2), f'{n1}->{n2}') #f' stands for F-String
                    for n1, n2 in G.edges]) #We use a dictionary to store all the values in the edges.
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels) #We draw the entire network

plt.show() #We display the graph
```
